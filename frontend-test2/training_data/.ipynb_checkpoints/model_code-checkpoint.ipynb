```python
gym_dataset = tf.keras.utils.image_dataset_from_directory(
    'dataset',
    image_size = (256,256),
    batch_size = 64,
    seed = 123,
    shuffle = True)
```

```python
dataset_classes = gym_dataset.class_names
print("Class names:", dataset_classes)
```

```python
#Visualizing images from the dataset after pre-processing

plt.figure(figsize = (14,14))
for images, labels in gym_dataset.take(4):
  for i in range(8):
    axis = plt.subplot(3, 4, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(dataset_classes[labels[i]])
    plt.axis("off")
```

```python
def get_gym_dataset_partitions_tf(data, training_split=0.8, validation_split=0.1, testing_split=0.1, shuffle=True, shuffle_size=7000):
    assert (training_split + testing_split + validation_split) == 1

    dataset_size = len(data)

    if shuffle:
        # Shuffle the entire dataset to ensure randomness
        dataset = data.shuffle(shuffle_size, seed=10)

    # Calculate the number of samples for each split
    train = int(training_split * dataset_size)
    validation = int(validation_split * dataset_size)

    # Create training, validation, and testing datasets
    train_dataset = data.take(train)    
    val_dataset = data.skip(train).take(validation)
    test_dataset = data.skip(train + validation)

    return train_dataset, val_dataset, test_dataset

# Usage example
train_dataset, val_dataset, test_dataset = get_gym_dataset_partitions_tf(gym_dataset)
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_dataset = train_dataset.cache().shuffle(800).prefetch(buffer_size=AUTOTUNE)
val_dataset = val_dataset.cache().shuffle(800).prefetch(buffer_size=AUTOTUNE)
test_dataset = test_dataset.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
#Data Augmentation Techniques

from tensorflow.keras import models, layers
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomZoom(0.1),
    layers.RandomRotation(0.1),
    layers.RandomContrast(0.1),
    layers.RandomBrightness(0.1),
    layers.RandomTranslation(0.1, 0.1),
    layers.GaussianNoise(0.02),
])
```

```python
train_dataset = train_dataset.map(
    lambda x, y: (data_augmentation(x, training=True), y)
).prefetch(buffer_size=AUTOTUNE)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
#Adding layers to the model

model = tf.keras.Sequential([
  tf.keras.layers.Rescaling(1./255, input_shape=(256, 256, 3)),
  tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
  tf.keras.layers.MaxPooling2D(2, 2),
    
  tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
  tf.keras.layers.MaxPooling2D(2,2),
    
  tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
  tf.keras.layers.MaxPooling2D(2,2),
    
  tf.keras.layers.Conv2D(256, (3,3), activation='relu'),
  tf.keras.layers.MaxPooling2D(2,2),
    
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(256, activation='relu'),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(5, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.001),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=['accuracy'])
```

```python
import time
from tensorflow import keras

timer_start = time.time()
```

```python
history = model.fit(
    train_dataset,
    batch_size = 64,
    validation_data=val_dataset,
    epochs=50
)
```

```python
time_trained = time.time() - timer_start
print (f"Time taken to train the model : {time_trained:.2f} seconds")
```

```python
#Plot Loss
plt.figure(figsize=(5, 5))
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Model Loss', fontsize=15)
plt.legend(['loss', 'validation_loss'],loc='upper right')
plt.show()

#Plot Accuracy
plt.figure(figsize=(5, 5))
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.title('Model Accuracy', fontsize=15)
plt.legend(['accuracy', 'validation_accuracy'],loc='lower right')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, classification_report

true_y_label = []
y_pred_all = []

for x_batch, y_batch in test_dataset:
    y_pred_probs = model.predict(x_batch) 
    y_pred_batch = np.argmax(y_pred_probs, axis=1)
    
    true_y_label.extend(y_batch.numpy())
    y_pred_all.extend(y_pred_batch)

true_y_label = np.array(true_y_label)
y_pred_all = np.array(y_pred_all)

#Calculate evaluation metrics values

accuracy = accuracy_score(true_y_label, y_pred_all)
conf_matrix = confusion_matrix(true_y_label, y_pred_all)
classification_report = classification_report(true_y_label, y_pred_all)
F1Score = f1_score(true_y_label, y_pred_all, average='weighted')

print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", classification_report)
print("F1 Score (Weighted):", F1Score)
```
